```python
import pandas as pd 
```

```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt
import random
```

```python
def initial ():
    m = 0
    c = 0
    return m,c
```

```python
def cal_cost (X,Y,m,c):
    n = len(Y)
    predict = m * X + c
    cost = np.sum((predict - Y) ** 2) /( 2 * n)
    return cost
```

```python
def cal_gradient(X,Y,m,c):
    n = len(Y)
    dm = np.sum(-X * (Y - (m*X + c))) / n
    dc = np.sum(-(Y - (m*X + c))) / n
    return dm, dc
```

```python
def update_parameters(m, c, dm, dc, learning_rate):
    m -= learning_rate * dm
    c -= learning_rate * dc
    return m, c
```

```python
def linear_regression(X, y, learning_rate, iterations):
    m, c = initial()
    for i in range(iterations):
        cost = cal_cost(X, y, m, c)
        dm, dc = cal_gradient(X, y, m, c)
        m, c = update_parameters(m, c, dm, dc, learning_rate)
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}")
    return m, c
```

```python
df = pd.read_csv('dataset\linear_train.csv')
```

```python
v1 = df['x']
v2 = df['y']
learning_rate = 0.00005
iterations = 1000
```

```python
m , c = linear_regression(v1,v2,learning_rate,iterations)
```
